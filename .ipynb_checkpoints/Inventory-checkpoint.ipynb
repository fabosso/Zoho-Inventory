```python
# Importar librería de pandas
import pandas as pd
```

```python
# Leer el .csv con todos los items de Zoho Inventory
items = pd.read_csv("Item.csv")
print(items)
```

```python
# Leer el .csv con el composite item descargado
composite = pd.read_csv("Composite_Item.csv")
print(composite)
```

```python
# Leer el Google Sheets con el listado de materiales (por ejemplo, el de Witech)
sheets_link = "https://docs.google.com/spreadsheets/d/e/2PACX-1vQNrd_norJhcPQQaRg_z8qek9y55mziVDbD061dd2GDaG7bRAOcF3Iz5dfYYAwDVR5hjDs3Jw3qmVAW/pub?output=xlsx"
materials = pd.read_excel(sheets_link)
print(materials)
```

Alternativamente, se puede leer desde local haciendo:
materials = pd.read_excel("Materiales.xlsx")

```python
# Obtener sólo las filas de Automation
items_automation = items[items['Purchase Account'] == 'Automation materials and equipment']
print(items_automation)
```

```python
# Obtener sólo las columnas que me interesan (nombre de item, SKU, y precio de venta)
items_automation = items_automation[['Item Name', 'SKU', 'Selling Price']]
print(items_automation)
```

```python
# Obtener sólo las columnas 1 ("Cantidad") y 3 ("Modelo", que va a ser lo mismo que "SKU"), a partir de la primera fila, y sin valores NA
materials_refined = materials.iloc[1:, [1,3]].dropna()
print(materials_refined)
```

```python
# Renombrar las columnas 3 (SKU) y 1 (Mapped Quantity), en ese orden
materials_refined.rename(columns={'Unnamed: 3':'SKU', 'Unnamed: 1':'Mapped Quantity'}, inplace=True)
print(materials_refined)
```

```python
# Filtrar para que sólo queden aquellas filas que tengan un SKU válido (existente en Zoho) y resetear el index de las filas.
materials_refined = materials_refined.loc[materials_refined['SKU'].isin(items_automation['SKU'])]
materials_refined.reset_index(inplace=True, drop=True)
print(materials_refined)
```

```python
# Crear una columna con los nombres de los mapped item names.
index = 0
new_dict = dict()
for sku in materials_refined['SKU']:
    item_name = items_automation.loc[sku == items_automation['SKU'], 'Item Name'].values[0]
    new_dict[index] = item_name
    index = index + 1
mapped_items = pd.DataFrame.from_dict({'Mapped Item Name': new_dict})
print(mapped_items)
```

```python
# Concatenar las columnas (eje vertical)
output = pd.concat([mapped_items, materials_refined], axis=1)
print(output)
```

```python
# Droppear la columna SKU, ya no la necesitamos
del output['SKU']
print(output)
```

```python
# Droppear las columnas viejas del composite, y concatenar materiales_refined.
del composite['Mapped Item Name']
del composite['Mapped Quantity']
output = pd.concat([composite, output], axis=1)
print(output)
```

```python
# Escribir a un nuevo archivo .csv
output.to_csv('zoho_composite_item.csv', index=False)
```
